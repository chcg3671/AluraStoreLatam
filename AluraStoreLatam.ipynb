#1. Análisis de facturación



calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
Ing_total1 = tienda["Precio"].sum()
Ing_total1
```

```python
Ing_total2 = tienda2["Precio"].sum()
Ing_total2
```

```python
Ing_total3 = tienda3["Precio"].sum()
Ing_total3
```

```python
Ing_total4 = tienda4["Precio"].sum()
Ing_total4
```

```python
print(f"🏪 Ingreso total Tienda 1: ${Ing_total1:,.2f}")
print(f"🏪 Ingreso total Tienda 2: ${Ing_total2:,.2f}")
print(f"🏪 Ingreso total Tienda 3: ${Ing_total3:,.2f}")
print(f"🏪 Ingreso total Tienda 4: ${Ing_total4:,.2f}")
```

```python
# Datos
Ingresos = [Ing_total1, Ing_total2, Ing_total3, Ing_total4]
Tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

mejor_ingreso = max(Ingresos)
peor_ingreso = min(Ingresos)
tienda_mejor_ing = Ingresos.index(mejor_ingreso) + 1
tienda_peor_ing = Ingresos.index(peor_ingreso) + 1

print(f"\n👍🏼 Tienda {tienda_mejor_ing} generó MÁS dinero\n")
print(f"👎🏼 Tienda {tienda_peor_ing} generó MENOS dinero")
```

```python
import matplotlib.pyplot as plt


# Crear gráfico de pastel
plt.figure(figsize=(7, 7))
plt.pie(Ingresos, labels=Tiendas, autopct='%1.1f%%', startangle=90, colors=['#66c2a5','#fc8d62','#8da0cb','#e78ac3'])
plt.suptitle('Ingresos Totales por Tienda', fontsize=14, fontweight='bold')
plt.title('La Tienda 4 fue la que menos dinero produjo', fontsize=10)
plt.axis('equal')  # Asegura que el gráfico sea circular
plt.show()
```

calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
tienda3['Calificación'].mean()
tienda3.groupby('Categoría del Producto').mean(numeric_only=True)
Cal_Prom3 = tienda3.groupby('Categoría del Producto')['Calificación'].mean()
Cal_Prom3
```

```python
tienda4['Calificación'].mean()
tienda4.groupby('Categoría del Producto').mean(numeric_only=True)
Cal_Prom4 = tienda4.groupby('Categoría del Producto')['Calificación'].mean()
Cal_Prom4
```

```python
# Calcular promedio de calificaciones por tienda
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

def mostrar_estrellas(calificacion):
    estrellas = int(round(calificacion))
    return "⭐" * estrellas + "☆" * (5 - estrellas)


# Mostrar resultados
print(f"🏪 Calificación promedio Tienda 1: {promedio_t1:.2f} {mostrar_estrellas(promedio_t1)}")
print(f"🏪 Calificación promedio Tienda 2: {promedio_t2:.2f} {mostrar_estrellas(promedio_t2)}")
print(f"🏪 Calificación promedio Tienda 3: {promedio_t3:.2f} {mostrar_estrellas(promedio_t3)}")
print(f"🏪 Calificación promedio Tienda 4: {promedio_t4:.2f} {mostrar_estrellas(promedio_t4)}")

calificaciones = {
    'Tienda 1': promedio_t1,
    'Tienda 2': promedio_t2,
    'Tienda 3': promedio_t3,
    'Tienda 4': promedio_t4
}

mejor_calif = max(calificaciones, key=calificaciones.get)
peor_calif = min(calificaciones, key=calificaciones.get)
print(f"\n👍🏼 {mejor_calif} tiene los clientes MÁS contentos\n")
print(f"👎🏼 {peor_calif} tiene los clientes MENOS contentos")
```
